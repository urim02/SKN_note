```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
```

## Scatter plots

데이터의 분포를 확인하기 위해서 사용

```python
np.random.seed(0)

x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))

sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

fig, ax = plt.subplots()
ax.scatter(x, y, s=sizes, c=colors)

plt.scatter(x, y, s = sizes, c = colors)
plt.show()
```

### Bar Charts

범주별 빈도 확인 시 사용

```python
x = 0.5 + np.arange(10)
y = np.random.uniform(2, 7, len(x))

fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor ='black', linewidth = 0.7)

ax.set(
    xlim = (0, 8), xticks = np.arange(1, 8),
    ylim = (0, 8), yticks = np.arange(1, 8)
)

plt.show()
```

### Line Chart

시간의 흐름에 따른 변화를 보고자 할 때 사용

```python
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

fig, ax = plt.subplots()

ax.plot(x, y, linewidth = 2.0)

ax.set(
    xlim = (0, 8), xticks = np.arange(1, 8), 
    ylim = (0, 8), yticks = np.arange(1, 8)
)

plt.show()
```

### Pie Charts

카페고리별 비율 시각화할 때 사용

```python
labels = 'Dogs', 'Hamsters', 'Cats', 'Frogs'
sizes = [45, 30, 15, 10]
explode = (0, 0.1, 0, 0)

fig1, ax1 = plt.subplots()

ax1.pie(
    sizes,
    explode = explode,
    labels = labels,
    autopct='%1.1f%%', # 퍼센트 값 표시
    shadow = True,
    startangle = 90
)

ax1.axis('equal')

plt.show()
```

### Histograms

데이터 분포를 확인하기 위해 사용

```python
x = 4 + np.random.normal(0, 1.5, 200)

fig, ax = plt.subplots()

ax.hist(x, bins = 8, linewidth = 0.5, edgecolor = 'black')

ax.set(
  xlim=(0, 8), xticks=np.arange(1, 8),
  ylim=(0, 56), yticks=np.linspace(0, 56, 9))

plt.show()
```

## Seaborn

### Scatter Plots

```python
tips = sns.load_dataset('tips')

# 중복값, 결측값 제거
tips = tips.drop_duplicates()
tips.isnull().sum(axis=0)

# Seaborn 플롯 함수들은 범주형 변수를 문자열로 처리
tips['size'] = tips['size'].astype(str)

# size 크기 순서대로 정렬
tips = tips.sort_values(by=['size'])
```

```python
tips
```

```python
sns.scatterplot(
    tips, 
    x = 'total_bill',
    y = 'tip',
    hue = 'size',
)
```

```python
sns.scatterplot(
    tips, 
    x = 'sex',
    y = 'tip',
    hue = 'size'
)
```
